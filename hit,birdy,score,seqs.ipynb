```python
import matplotlib.pyplot as plt
import xlrd
import os
import math
import xlrd ,xlwt
import pandas as pd

#var!
data_name='coordinatebHSBC53.xls'
final_name='HSBC53.xls'

#read        
file_path = os.path.dirname("C://Users/lilly/Desktop/HiSeqGan-master/")
base_path = os.path.join(file_path, data_name)
book = xlrd.open_workbook(base_path)
sheet1 = book.sheets()[0]

#write 
finishfold= ''
folderpath = 'C://Users/lilly/Desktop/HiSeqGan-master/'      
booknew = xlwt.Workbook(encoding='utf-8',style_compression=0)
sheet2 = booknew.add_sheet('cordination',cell_overwrite_ok=True) 

#calculate
df = pd.read_excel('C://Users/lilly/Desktop/HiSeqGan-master/'+data_name)

#folder
if os.path.exists(folderpath):
    print('Warning : folder is exist.')
else:
    print('Creating folder ....')
    try:
        os.makedirs(folderpath)
        print('Success to create  folder.')
    except Exception as e:
        print('Failed to create folder')
        
#new column
sheet2.write(0, 16, "HIT")
sheet2.write(0, 17,"ball") #球種欄
sheet2.write(0, 18,"score(上：下)")

#raw data
for i in range(len(df)+1):
    sheet2.write(i, 0, sheet1.cell(i, 0).value)
    sheet2.write(i, 1, sheet1.cell(i, 1).value)
    sheet2.write(i, 2, sheet1.cell(i, 2).value)
    sheet2.write(i, 3, sheet1.cell(i, 3).value)
    sheet2.write(i, 4, sheet1.cell(i, 4).value)
    sheet2.write(i, 5, sheet1.cell(i, 5).value)
    sheet2.write(i, 6, sheet1.cell(i, 6).value)
    sheet2.write(i, 7, sheet1.cell(i, 7).value)
    sheet2.write(i, 8, sheet1.cell(i, 8).value)
    sheet2.write(i, 9, sheet1.cell(i, 9).value)
    sheet2.write(i, 10, sheet1.cell(i, 10).value)
    sheet2.write(i, 11, sheet1.cell(i, 11).value)
    sheet2.write(i, 12, sheet1.cell(i, 12).value)
    sheet2.write(i, 13, sheet1.cell(i, 13).value)
    sheet2.write(i, 14, sheet1.cell(i, 14).value)
    sheet2.write(i, 15, sheet1.cell(i, 15).value)
    booknew.save(folderpath+final_name)

d=0  #判斷d012
    
#球位置
mind1=df[1:]['ry下面的人'].min()
maxd1=df[1:]['ry下面的人'].max()
d1_40=mind1+(maxd1-mind1)*0.4
d1_60=mind1+(maxd1-mind1)*0.55
#d=2
print(d1_40)
print(d1_60)


mind2=df[1:]['by上面的人'].min()
maxd2=df[1:]['by上面的人'].max()
d2_40=mind2+(maxd2-mind2)*0.45
d2_60=mind2+(maxd2-mind2)*0.6

print(d2_40)
print(d2_60)
#d=1


for i in range(1,len(df)+1):
    d=sheet1.cell(i, 15).value
    if(d==1):
        position=sheet1.cell(i, 12).value
        if(position<d2_40):
            sheet2.write(i, 16,"B")
        elif(position>d2_60):
            sheet2.write(i, 16,"F")
        else:
            sheet2.write(i, 16,"M")
    elif(d==2):
        position=sheet1.cell(i, 6).value
        if(position<d1_40):
            sheet2.write(i, 16,"F")
        elif(position>d1_60):
            sheet2.write(i, 16,"B")
        else:
            sheet2.write(i, 16,"M")
    elif(d==0 and sheet1.cell(i-1, 15).value !=0 and sheet1.cell(i-1, 15).value !="d"):
        if(sheet1.cell(i-1, 15).value==1):#前一個是1
            position=sheet1.cell(i, 6).value#寫2的位置
            if(position<d1_40):
                sheet2.write(i, 16,"F")
            elif(position>d1_60):
                sheet2.write(i, 16,"B")
            else:
                sheet2.write(i, 16,"M")
        else:#前一個是2
            position=sheet1.cell(i, 12).value#寫2的位置
            if(position<d2_40):
                sheet2.write(i, 16,"B")
            elif(position>d2_60):
                sheet2.write(i, 16,"F")
            else:
                sheet2.write(i, 16,"M")
    else:
        sheet2.write(i, 16,"")
    booknew.save(folderpath+final_name)

print("ok")

```

```python
#戴營戴發球 戴營對發球 看下一局的第一個分類上一個序列
#戴書戴發球 戴書對發球

import matplotlib.pyplot as plt
import os
import math
import xlrd ,xlwt
import pandas as pd

#variable!
data_name='HSBC53.xls'
final_test1='HSBC53_11.xls'#11 1win
final_test2='HSBC53_12.xls'#12 1lose
final_test3='HSBC53_21.xls'#21 2lose
final_test4='HSBC53_22.xls'#22 2win

#read        
file_path = os.path.dirname("C://Users/lilly/Desktop/HiSeqGan-master/")
base_path = os.path.join(file_path, data_name)
book = xlrd.open_workbook(base_path)
sheet1 = book.sheets()[0]
# book = xlrd.open_workbook(base_path)

#write 
# finishfold= 'txttest/'#資料夾名
folderpath = 'C://Users/lilly/Desktop/HiSeqGan-master/0510test/'      
 

#folder
if os.path.exists(folderpath):
    print('Warning : folder is exist.')
else:
    print('Creating folder ....')
    try:
        os.makedirs(folderpath)
        print('Success to create  folder.')
    except Exception as e:
        print('Failed to create folder')


df = pd.read_excel("C://Users/lilly/Desktop/HiSeqGan-master/"+data_name)

#
player=0 #發球者是誰
temp=False #有無紀錄發球者
state=0 #贏球是誰
size=len(df)+1

indexstart=0#開頭index
indexend=0#末端index
inEND=False #有無紀錄末端index

num11=1
booknew1 = xlwt.Workbook(encoding='utf-8',style_compression=0)
sheet2 = booknew1.add_sheet('cordination',cell_overwrite_ok=True)
sheet2.write(0, 0,"seq")

num12=1
booknew2 = xlwt.Workbook(encoding='utf-8',style_compression=0)
sheet3 = booknew2.add_sheet('cordination',cell_overwrite_ok=True)
sheet3.write(0, 0,"seq")

num21=1
booknew3 = xlwt.Workbook(encoding='utf-8',style_compression=0)
sheet4 = booknew3.add_sheet('cordination',cell_overwrite_ok=True)
sheet4.write(0, 0,"seq")

num22=1
booknew4 = xlwt.Workbook(encoding='utf-8',style_compression=0)
sheet5 = booknew4.add_sheet('cordination',cell_overwrite_ok=True)
sheet5.write(0, 0,"seq")

for i in range(1,len(df)+1):
    d=sheet1.cell(i, 15).value #d
    if(d==0 and (temp is False)):
        continue
    elif(d==1 and (temp is False)):
        player=1
        temp=True
        indexstart=i
    elif(d==2 and (temp is False)):
        player=2
        temp=True
        indexstart=i
    elif(d==0 and (temp is True)and sheet1.cell(i, 16).value==""):
        if(inEND is False):
            indexend=i
            inEND=True
        if((i+1) == (size)):
            break
        elif((sheet1.cell(i+1, 15).value!=0)):
            state=sheet1.cell(i+1, 15).value
            #將序列加進去 紀錄頭尾index ep 根據index取ball欄位的值(原檔案)存到新檔案 分清楚新舊index
            if((player == 1)and (state ==1)):
                for k in range(indexstart,indexend):#k是新檔案的index 
                    sheet2.write(num11, 0,sheet1.cell(k, 17).value)
                    num11+=1
                sheet2.write(num11, 0,"end")
                num11+=1
                booknew1.save(folderpath+final_test1)
                
            elif((player == 1)and (state == 2)):
                for k in range(indexstart,indexend):#k是新檔案的index 
                    sheet3.write(num12, 0,sheet1.cell(k, 17).value)
                    num12+=1
                sheet3.write(num12, 0,"end")
                num12+=1
                booknew2.save(folderpath+final_test2)
                
            elif((player == 2)and (state == 1)):
                for k in range(indexstart,indexend):#k是新檔案的index 
                    sheet4.write(num21, 0,sheet1.cell(k, 17).value)
                    num21+=1
                sheet4.write(num21, 0,"end")
                num21+=1 
                booknew3.save(folderpath+final_test3)
                
            elif((player == 2)and (state == 2)):
                for k in range(indexstart,indexend):#k是新檔案的index 
                    sheet5.write(num22, 0,sheet1.cell(k, 17).value)
                    num22+=1 
                sheet5.write(num22, 0,"end")
                num22+=1 
                booknew4.save(folderpath+final_test4)
                
            temp=False 
            inEND=False
        continue
         
```
