# 攻防圓餅圖

```python
#1,2,3種組合 分3種
import pandas as pd
import xlrd ,xlwt
import numpy
# #讀        
data_name='seqGAN_output_hs5TLL_0608_type2.xls'


df = pd.read_excel('C://Users/lilly/Desktop/HiSeqGan-master/seqganoutput0608/'+data_name)


attack=0
defence=0
others=0
for i in range(0,100):
    rownum=numpy.sum(df[i:i+1].count(axis =1))
    #
    for j in range(1,rownum):

        tag1=df.iloc[i,j-1]
        tag2=df.iloc[i,j] 

        if(((tag1 == 'Clear')or(tag1 == 'Drive')or(tag1 == 'Net_lift')or(tag1 == 'Dribble')or(tag1 == 'Defensive_short'))and((tag2 == 'Clear')or(tag2 == 'Drive')or(tag2 == 'Net_lift')or(tag2 == 'Dribble')or(tag2 == 'Defensive_short'))):
             defence=defence+1
        elif(((tag1 == 'Net_smash')or(tag1 == 'Drop')or(tag1 == 'Smash')or(tag1 == 'Net'))and((tag2 == 'Net_smash')or(tag2 == 'Drop')or(tag2 == 'Smash')or(tag2 == 'Net'))):    
             attack=attack+1
        else:
             others=others+1
                

all=attack+defence+others
# print(defence)
alldict={}
dict={'Attack':round(attack/all,3)}
alldict={**alldict,**dict}
dict={'Defense':round(defence/all,3)}
alldict={**alldict,**dict}
dict={'Combined':round(others/all,3)}
alldict={**alldict,**dict}
alldict
```
